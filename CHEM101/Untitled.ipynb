```python
import numpy as np
import matplotlib.pyplot as plt

molarity = np.array([0.500, 0.200, 0.100])
density = np.array([1.066, 1.014, 1.001])

coefficients = np.polyfit(molarity, density, 1)  # slope, intercept
slope, intercept = coefficients

y_fit = slope * molarity + intercept

ss_res = np.sum((density - y_fit)**2)
ss_tot = np.sum((density - np.mean(density))**2)
r_squared = 1 - ss_res / ss_tot

plt.scatter(molarity, density, label="_nolegend_")
plt.plot(molarity, y_fit, label=f'Fit: y={slope:.6f}x + {intercept:.6f}\n$R^2$={r_squared:.5f}')
plt.xlabel('Molarity (mol/L)')
plt.ylabel('Density (g/cm³)')
plt.title('Standard Density of NaCl Solution vs Molarity')
plt.legend()
plt.show()
```
