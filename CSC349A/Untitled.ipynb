```python
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
srate = 10000
g = 9.81

def main(g, m, c):
    lis = np.zeros(int(srate))
    alpha = (g*m)/c
    beta = (c)/m
    indices = range(0, srate)
    times = np.array(indices)/srate
    for (a,b) in zip(indices, times):
        lis[a] = alpha*(1-np.exp(-beta*b))
        print(lis[a])
    return lis
        
lis = main(g, m = 1, c = 2)

fig = plt.figure()
ax = fig.add_axes((0.1,0.2,0.8,0.7))
ax.plot(range(0, 10000), lis)
plt.show()
print(lis)
```
